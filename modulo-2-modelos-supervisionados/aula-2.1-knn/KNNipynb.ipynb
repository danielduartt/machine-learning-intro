# Distância Euclidiana

```python
#       massa, cor
frutaA = [5, 0.75]
frutaB = [2, 0.50]

# aplicando a dist euclidiana
soma_dos_quadrados = 0
for(i, j) in zip(frutaA, frutaB):
    soma_dos_quadrados += ((i-j)**2)
resultado = soma_dos_quadrados**0.5
print(resultado)
```

```python
# Usando o numpy
import numpy as np

frutaA = np.array([5, 0.75])
frutaB = np.array([2, 0.50])

distancia = np.linalg.norm(frutaA - frutaB)
print(distancia)
```

# KNN para classificação


```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
data = pd.read_table('fruit_data_with_colors.txt')
data.head()
```

```python
x = data[["mass", "width", "height", "color_score"]] # Variáveis numéricas
y = data["fruit_label"] # Variável categórica
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_test, y_test)
```

```python
# Nesse caso os dados numéricos podem variar bastante, pois há frutas com massas bem pequenas e outras com massas bem maiores
# a Ordem de grandeza vai fazer diferença na distancia
# Grandezas bem diferentes, a grandeza maior vai influenciar muito mais do que as grandezas menores. Nesse necessitando de uma balanceamente para melhorar a performace dos nosso dados
```

## Preprocessando os Dados para o KNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
x_train = mm.fit_transform(x_train)
x_test = mm.fit_transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

# KNN para Regressão

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_regressor = KNeighborsRegressor(n_neighbors=3)
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
data_train, data_test, target_train, target_test = train_test_split(data, target, test_size=0.25)
```

```python
knn_regressor.fit(data_train, target_train)
```

```python
knn_regressor.score(data_test, target_test)
```
